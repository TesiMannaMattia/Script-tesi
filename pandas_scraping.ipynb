# Scraping di tabelle

Per effettuare scraping su dati contenuti in una tabella, invece di scrivere uno scraper ad hoc, si può utilizzare direttamente il pacchetto __Pandas__.  
Questo metodo può tornare utile nelle pagine con molte tabelle, ad esempio Wikipedia.

La sintassi è molto semplice, tramite **request.get()** si ottengono :
 - Gli html grezzi ed altre informazioni dal sito.
 - Altre informazioni.

Successivamente con il comando **.text** applicato ai dati ottenuti tramite **request.get()** si riporta la sintassi html nel suo stato originale su cui è possibile lavorare.

Con il comando  **pd.read_html()** si ottiene una lista nella quale sono presenti le tabelle.

```python
import pandas as pd
import requests
raw_data=requests.get('https://en.wikipedia.org/wiki/List_of_United_States_cities_by_area')
tabs = pd.read_html(raw_data.text)
tbl = tabs[1]
tbl
```

Si dimostri ciò che è stato detto in precedenza :
- Si osservi come faccia  **pd.read_html()**  ottenere una lista
- Si osservi la lunghezza della lista, significa che sono state ottenute due tabelle.

```python
print(type(tabs))
print(len(tabs))
```

## Un altro metodo
Un altro metodo equivalente per ottenere i dati è quello di inserire il link del sito
direttamente nel comando **pd.read_html()**.

```python
tabs = pd.read_html("https://en.wikipedia.org/wiki/List_of_United_States_cities_by_area")
tbl = tabs[1]
print(type(tabs))
tbl
```

### Esempio
